```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun(a, b, c, size):
    U = np.random.rand(size)
    
    samples = np.zeros(size)
    
    for i in range(size):
        if U[i] <= (c - a) / (b - a):
            samples[i] = a + np.sqrt(U[i] * (b - a) * (c - a))
        else:
            samples[i] = b - np.sqrt((1 - U[i]) * (b - a) * (b - c))
    
    return samples
```

```python
a = 0
b = 25
c = 20
size = 1000000

samples = fun(a, b, c, size)
```

```python
plt.hist(samples, bins = 100, density = True)
plt.show()
```
